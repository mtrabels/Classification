### Continuous variables

```python
from matplotlib import pyplot as plt
import numpy as np
import seaborn as sns
```

```python
def show_distrib(df, col, y):
    fig = plt.figure()
    fig, ax = plt.subplots(nrows=1, ncols=2,  sharex=False, sharey=False, figsize=((10, 5)))
    fig.suptitle(str(col)+"   vs   "+str(y), fontsize=20)
    ### distribution
    ax[0].title.set_text('density')
    for i in df[y].unique():
        sns.distplot(df[df[y]==i][col], hist=False, label=i, ax=ax[0])
    ax[0].legend()
    ax[0].grid(True)
   
    ### frequencies
    ax[1].title.set_text('bins')
    breaks = np.quantile(df[col], q=np.linspace(0,1,11))
    tmp = df.groupby([y, pd.cut(df[col], breaks, duplicates='drop')]).size().unstack().T #cut variable into groups and calculate frequence
    tmp = tmp[df[y].unique()] #take columns 0 and 1
    tmp["tot"] = tmp.sum(axis=1)
    for var in tmp.drop("tot", axis=1).columns:
         tmp[var] = tmp[var] / tmp["tot"]
    tmp.drop("tot", axis=1).plot(kind='bar', stacked=True, ax=ax[1], legend=False, grid=True)
    
    ### boxplot   
    g = sns.catplot(x=y, y=col, data=df_train, kind="box")
    g.fig.suptitle("Outliers")
    plt.show()
```

### Categorical variables

For categorical variables we can plot barplots

```python
def plot_cat(df, col, y):
    fig = plt.figure()
    fig.suptitle(str(col)+"   vs   "+str(y), fontsize=20)
    
    ### percentage
    ax = fig.add_subplot(111)
    ax.title.set_text('percentage')
    a = df.groupby(col)[y].count().reset_index()
    a = a.rename(columns={y:"tot"})
    b = df.groupby([col,y])[y].count().to_frame()
    b = b.rename(columns={y:0}).reset_index()
    b = b.merge(a, how="left")
    b["%"] = b[0] / b["tot"] *100
    sns.barplot(x=col, y="%", hue=y, data=b)
    ax.grid(True)
    
    ### count
    order = df.groupby(col)[y].count().index.tolist()
    g1 = sns.catplot(x=col, hue=y, data=df, kind='count', order=order)
    g1.fig.suptitle("count")
    
    plt.show()
```

### Statistics

To do this properly, we need to do the same observations on all variables. A good way is to perform statistic tests to make conclusions about dependencies of features with target. For continuous variables, a one way ANOVA test is computed and for categorical variables a chi-square test is done.

#### Categorical variables

```python
from scipy.stats import chi2_contingency
```

```python
for col in cat_columns:
    cont_table = pd.crosstab(index=df_train[col], columns=df_train[y])
    stat, p, dof, expected = chi2_contingency(cont_table)
    if p < 0.05:
        print("Column " + str(col) + " and column " + str(y) + " are statistically dependent", p)
    else:
        print("Column " + str(col) + " and column " + str(y) + " are statistically independent", p)
```

These features are interesting to keep for modeling since they might be discriminative.

#### Continuous variables

```python
import statsmodels.formula.api as smf
import statsmodels.api as sm
```

```python
df_tmp = df_train.copy()
df_tmp.columns = ["Col" + str(i) for i in df_tmp.columns]
```

```python
for col in cont_columns:
    model = smf.ols("Col" + str(col)+' ~ '+ "Col" + str(y), data=df_tmp).fit()
    table = sm.stats.anova_lm(model)
    p = table["PR(>F)"][0]
    coeff, p = None, p
    if p < 0.05:
        print("Column " + str(col) + " and column " + str(y) + " are correlated", p)
    else:
        print("Column " + str(col) + " and column " + str(y) + " are not correlated", p)
```

All the continuous variables seem to be different for each group.

We can also compute correlations between features to see if they are linearly related. If so, we can discard some features.

First we need to encode our categorical variables.

```python
df_train_trf = df_train.copy()
for col in cat_columns:
    df_train_trf[col] = df_train_trf[col].astype("category")
    df_train_trf[col] = df_train_trf[col].cat.codes
```

```python
fig, ax = plt.subplots(figsize=(20,20))  
corr_matrix = df_train_trf.corr(method="pearson")
sns.heatmap(corr_matrix, vmin=-1., vmax=1., annot=True, fmt='.2f', cmap="YlGnBu", cbar=True, linewidths=0.5, ax=ax)
plt.title("pearson correlation")
```

We can look at only one part of the matrix since it is symmetric. We can clearly observe at least 6 groups from this correlation matrix (correlation above 0.7):
- 24: {25, 26, 28} (migration) (keep 24, lowest p-value)
- 22: 23 (detailed houselhold) (keep 22)
- 37: 30 (family members u18 and veterans benefits) (keep 37)
- 31: 32 (keep 31, lowest p-value)
- 33: 34 (keep 33, lowest p-value)
- 38: 2 (industry recode and weeks worked in year) (keep 38)

Another interesting note is that the inserted column of the original dataset (24 -> 41) is not correlated with any other feature so we don't know what it corresponds to. Thus it is safer to discard it.

# Modeling

We need to find wether a person makes more or less than 50000/year so this is a binary classification problem.

We first try to train our model on all our features.

Let's split our trainset in train and val data. We have a lot of data so we can split it once by 80% train and 20% val.

## Random forest

Now let's train a random forest classifier. This classifier has the reputation to not overfit the data and gives good results without spending a lot of time tuning hyperparameters. Moreover, it can give important features used for classification.

### Random forest with all the features

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
clf = RandomForestClassifier(min_samples_leaf=1, max_features=1.0, max_samples=None, oob_score=True, n_jobs=-1, random_state=42)
```

Test model

Our accuracy is good, but we also need the fscore for the detailed performances of both categories.

```python
from sklearn.metrics import precision_recall_fscore_support
```

```python
y_pred = clf.predict(X_val)
precision, recall, fscore, _ = precision_recall_fscore_support(y_val, y_pred)
print("+50k precision:", precision[1])
print("+50k recall:", recall[1])
print("+50k fscore:", fscore[1])
print()
print("-50k precision:", precision[0])
print("-50k recall:", recall[0])
print("-50k fscore:", fscore[0])
```

We have very good results in detecting -50k category. However, the fscore of the +50k category is quite low. It is not surprising as we are in an imbalanced case.

### Feature selection

```python
columns_selected = [i for i in df_train_trf.columns if i not in [16, 17, 40, 41, 25, 26, 28, 23, 30, 32, 34, 2]]
X_train, X_val, y_train, y_val = train_test_split(df_train_trf[columns_selected], df_train_trf[y], test_size=0.2, random_state=42, shuffle=True, stratify=df_train_trf[y])
clf = RandomForestClassifier(min_samples_leaf=1, max_features=1.0, max_samples=None, oob_score=True, n_jobs=-1, random_state=42).fit(X_train, y_train)
y_pred = clf.predict(X_val)
precision, recall, fscore, _ = precision_recall_fscore_support(y_val, y_pred)
print("+50k precision:", precision[1])
print("+50k recall:", recall[1])
print("+50k fscore:", fscore[1])
print()
print("-50k precision:", precision[0])
print("-50k recall:", recall[0])
print("-50k fscore:", fscore[0])
```

### Remove Outliers

Let's try to see if we can remove outliers from features distributions of both categories but we need to first split categories into 2 dataframes and analyze the distributions of the different features.

```python
def find_outliers(df, col):
    df_plus = df_train[df_train[y]==1]
    df_minus = df_train[df_train[y]==0]
    fig, ax = plt.subplots(nrows=2, ncols=2,  sharex=False, sharey=False, figsize=(10, 5))
    fig.suptitle(col, fontsize=20)

    ### distribution minus 50
    ax[0][0].title.set_text('distribution -50')
    variable = df_minus[col]
    breaks = np.quantile(variable, q=np.linspace(0, 1, 11))
    variable = variable[ (variable > breaks[0]) & (variable < 
                        breaks[10]) ]
    sns.distplot(variable, hist=True, kde=True, kde_kws={"shade": True}, ax=ax[0][0])
    des = df_minus[col].describe()
    ax[0][0].axvline(des["25%"], ls='--')
    ax[0][0].axvline(des["mean"], ls='--')
    ax[0][0].axvline(des["75%"], ls='--')
    ax[0][0].grid(True)
    des = round(des, 2).apply(lambda x: str(x))
    box = '\n'.join(("min: "+des["min"], "25%: "+des["25%"], "mean: "+des["mean"], "75%: "+des["75%"], "max: "+des["max"]))
    ax[0][0].text(0.95, 0.95, box, transform=ax[0][0].transAxes, fontsize=10, va='top', ha="right", bbox=dict(boxstyle='round', facecolor='white', alpha=1))

    ### distribution plus 50
    ax[0][1].title.set_text('distribution +50')
    variable = df_plus[col]
    breaks = np.quantile(variable, q=np.linspace(0, 1, 11))
    variable = variable[ (variable > breaks[0]) & (variable < 
                        breaks[10]) ]
    sns.distplot(variable, hist=True, kde=True, kde_kws={"shade": True}, ax=ax[0][1])
    des = df_plus[col].describe()
    ax[0][1].axvline(des["25%"], ls='--')
    ax[0][1].axvline(des["mean"], ls='--')
    ax[0][1].axvline(des["75%"], ls='--')
    ax[0][1].grid(True)
    des = round(des, 2).apply(lambda x: str(x))
    box = '\n'.join(("min: "+des["min"], "25%: "+des["25%"], "mean: "+des["mean"], "75%: "+des["75%"], "max: "+des["max"]))
    ax[0][1].text(0.95, 0.95, box, transform=ax[0][1].transAxes, fontsize=10, va='top', ha="right", bbox=dict(boxstyle='round', facecolor='white', alpha=1))

    ### boxplot minus 50
    ax[1][0].title.set_text('outliers (log scale)')
    tmp_df = pd.DataFrame(df_minus[col])
    tmp_df[col] = np.log(tmp_df[col])
    tmp_df.boxplot(column=col, ax=ax[1][0])

    ### boxplot plus 50
    ax[1][1].title.set_text('outliers (log scale)')
    tmp_df = pd.DataFrame(df_plus[col])
    tmp_df[col] = np.log(tmp_df[col])
    tmp_df.boxplot(column=col, ax=ax[1][1])

    plt.show()
```

### GradientBoostingClassifier

Gradient boosting classifier has also the reputation of being performant and fast. Also, Gradient Boosting is a sequential process and thus every time it makes an incorrect prediction, it focuses more on that incorrectly predicted data point. So, if the first iteration gives an accuracy of 80 %, the second iteration would focus on the remaining 20%.

```python
from sklearn.ensemble import GradientBoostingClassifier
```

```python
clf_boost= GradientBoostingClassifier(max_features=1.0, n_estimators=1000, random_state=42, learning_rate=0.1).fit(X_train, y_train)
```

```python
y_pred = clf_boost.predict(X_val)
precision, recall, fscore, _ = precision_recall_fscore_support(y_val, y_pred)
print("+50k precision:", precision[1])
print("+50k recall:", recall[1])
print("+50k fscore:", fscore[1])
print()
print("-50k precision:", precision[0])
print("-50k recall:", recall[0])
print("-50k fscore:", fscore[0])
```

Gradient boosting with 1000 estimators seem to give better results (+3% fscore) than random forest. This is mainly due to the large increase of precision (from 69% to 75%) so we have much less false positives.
